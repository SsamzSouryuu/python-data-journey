```python
#percobaan kedua
brix_list = [22,14,18,10,25, 20]

for brix in brix_list:
  if brix >= 20:
    print(f"Sampel memiliki Grade: A {brix}")
  elif brix >= 15:
    print(f"Sampel memiliki Grade: B {brix}")
  else:
    print(f"Sampel memiliki Grade: C {brix}")
```
