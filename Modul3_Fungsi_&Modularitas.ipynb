```python
#Percobaan 2

def hitung_kadar_air(berat_basah, berat_kering):
  kadar_air = ((berat_basah - berat_kering) / berat_basah) * 100
  return kadar_air

hasil = hitung_kadar_air(100, 48)
print (hasil)

def klasifikasi_brix(hasil_klasifikasi): #ga kayak kadar air, returnnya udh ada di if/elif/else.
  if hasil_klasifikasi >= 20:
    return "SS"

  elif hasil_klasifikasi >= 15:
    return "A"

  elif hasil_klasifikasi >= 10:
    return "B"

  else:
    return "C"
#Sekali ngitung doang, nanti tinggal kasih f-string sama angka brix muncul.
hasil2 = klasifikasi_brix(20)
print (hasil2)

print (f"Kadar Air: {hasil}%, Grade Brix: {hasil2}")
```
